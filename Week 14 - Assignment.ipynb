```python
import matplotlib.pyplot as plt

# Data biaya
labels = ['Prevention', 'Appraisal', 'Internal Failure', 'External Failure']
costs_without_sqa = [0, 50_000_000, 300_000_000, 800_000_000]
costs_with_sqa = [75_000_000, 50_000_000, 297_500_000, 210_000_000]

# Posisi sumbu X
x = range(len(labels))
bar_width = 0.35

# Membuat figure dan axes
fig, ax = plt.subplots(figsize=(10, 6))

# Membuat bar chart
bars1 = ax.bar([i - bar_width/2 for i in x], costs_without_sqa, width=bar_width, label='Tanpa SQA')
bars2 = ax.bar([i + bar_width/2 for i in x], costs_with_sqa, width=bar_width, label='Dengan SQA')

# Menambahkan label dan judul
ax.set_xlabel('Komponen Biaya')
ax.set_ylabel('Biaya (Rp)')
ax.set_title('Perbandingan Cost of Software Quality (CoSQ)')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Menata layout dan menampilkan
plt.tight_layout()
plt.show()
```
